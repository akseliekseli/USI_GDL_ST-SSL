```python
###imports

#import torch  ##2.0.0+cu118
import pandas as pd
import numpy as np
import os
from pylab import *
import matplotlib.pyplot as plt

from glu_tf import GLU #for the gated mechnaism - GLU
```

```python
#Path to the target csv file. This notebook should be in the same folder as the csv-data file 11_Autonomous_Systems.csv
path = os.getcwd()
print(path)
```

```python
###scanNet, shapeNet, SunRGB-D, A2D2
datasetBikeOne = np.load(path+'./data/NYCBike1/train.npz')
for file in datasetBikeOne.files:
    print(file, datasetBikeOne[file].shape)

```

```python
x = datasetBikeOne["x"]
y = datasetBikeOne["y"]
```

```python
plt.figure(figsize=(18, 18))
plt.imshow(x[0, :, :, 0], cmap='jet', aspect='auto') #
plt.xlabel('timeslots')
```

```python
# Assuming x is your 4D tensor of shape (#timeslots, #lookback_window, #nodes, #flow_types)
plt.figure(figsize=(18, 18))
flow_type = 0  # Assuming you want to plot the first flow type
x_agg = np.mean(x[:, :, :, flow_type], axis=1)  # Aggregate along the flow type and lookback window dimensions
x_heatmap = np.transpose(x_agg)  # Transpose to match the heatmap orientation
plt.imshow(x_agg, cmap='hot')
plt.show()
```

```python

plt.figure(figsize=(10, 10))
#to show all the data
### there are only 2 flow type, 0 and 1.
data = x[0, :, :,0]

# Plot the data
plt.plot(data, label='flow')
plt.xlabel('timeslots')
plt.ylabel('Y-axis')
plt.title("Traffic Flow Graph")
plt.show()
plt.legend(framealpha=1, frameon=True);
```
