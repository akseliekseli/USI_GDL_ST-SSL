Visualisations experiments to understand the dataset more

```python
###imports
import pandas as pd
import numpy as np
import os
from pylab import *
import matplotlib.pyplot as plt
```

```python
#Path to the target csv file. This notebook should be in the same folder as the csv-data file 11_Autonomous_Systems.csv
path = os.getcwd()
print(path)
```

```python
###scanNet, shapeNet, SunRGB-D, A2D2
datasetBikeOne = np.load(path+'./data/NYCBike1/train.npz')
for file in datasetBikeOne.files:
    print(file, datasetBikeOne[file].shape)

```

```python
x = datasetBikeOne["x"]
y = datasetBikeOne["y"]
```

```python
##comparison between the 2 flows in the dataset

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(
    #x[:, 0],
    #y[:, 0],
    x[0, :, :, 0],
    x[0, :, :, 1],
    c="blue",
    alpha=0.9,
)
ax.set(
    aspect="equal",
    xlabel="$X_1$",
    ylabel="$X_2$",
 )
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(x[0, :, :, 0], cmap='gnuplot', aspect='auto') #
plt.xlabel('nodes') #maybe
plt.ylabel('timeslots')
```

```python
###visualising the data in the x_offsets of dataset
plt.plot(datasetBikeOne["x_offsets"])
plt.xlabel('timeslots')
plt.ylabel('x offsets')
plt.title("X Offsets object in NCBike1 dataset")
plt.show()
```

```python
###visualising the data in the x_offsets of dataset
###didn't show anything like the x_offset dataset
plt.plot(datasetBikeOne["y_offsets"])
plt.xlabel('timeslots')
plt.ylabel('y offsets')
plt.title("Y Offsets object in NYCBike1 dataset")
plt.show()
```

```python
##reason for the y_offsets not producing any visualisation is that it only has one array

yOffShape = datasetBikeOne["y_offsets"].shape
yOff = datasetBikeOne["y_offsets"]
print(f"this is the array size {yOffShape}")
print(f"this is the array of the y offset: {yOff} ")
```

```python
###visualising the data in the x of dataset with flow 0
plt.figure(figsize=(10, 10))
#to show all the data
### there are only 2 flow type, 0 and 1.
data = x[0, :, :,0]

# Plot the data
plt.plot(data, label='flow')
plt.xlabel('timeslots')
plt.ylabel('Y-axis')
plt.title("Traffic Flow Graph")
plt.show()
```

```python
###visualising the data in the x of dataset with flow 1
plt.figure(figsize=(10, 10))
#to show all the data
### there are only 2 flow type, 0 and 1.
data = x[0, :, :,1]

# Plot the data
plt.plot(data, label='flow')
plt.xlabel('timeslots')
plt.ylabel('Y-axis')
plt.title("Traffic Flow Graph")
plt.show()
```

```python
##comparing the different flow types using all nodes in the dataset for time for timeslot 0

plt.figure(figsize=(8, 8))
#to show all the data
####so the maximum for each point in the x structure is x[2999, ?, 127, 1]
line1,  = plt.plot(x[0, 0, :, 0], label='flow 1')
line2,  = plt.plot(x[0, 0, :, 1], label='flow 2')
# Plot the data
#plt.plot(data)
plt.xlabel('nodes')
plt.ylabel('flow type values')
plt.title("Traffic Flow Graph at timeslot 0 in NYCBike1 dataset - X Object")
plt.legend(handles=[line1, line2])
plt.show()
```

```python
##comparing the different flow types using all nodes in the dataset for time for timeslot 0

plt.figure(figsize=(8, 8))
#to show all the data
####so the maximum for each point in the y structure is y[2999, 0, 127, 1]
#or the last one flow type, you need to put :0 or it wont visualise anything
line1,  = plt.plot(y[0, 0, :, 0], label='flow 1')
line2,  = plt.plot(y[0, 0, :, 1], label='flow 2')
# Plot the data
#plt.plot(data)
plt.xlabel('nodes')
plt.ylabel('flow type values')
plt.title("Traffic Flow Graph at timeslot 0 in NYCBike1 dataset - Y Object")
plt.legend(handles=[line1, line2])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
 
# Change the Size of Graph using
# Figsize
fig = plt.figure(figsize=(10, 10))
 
# Generating a 3D sine wave
ax = plt.axes(projection='3d')

#x , y = io[:,0], io[:,1]
x = datasetBikeOne["x"]
y = np.sin(x)
z = y*np.sin(x)
c = x + y
ax.scatter(x, y,c)
 
# turn off/on axis
plt.axis('on')
plt.xlabel('nodes')
plt.ylabel('lookback_window')
ax.set_zlabel('nodes')
# show the graph
plt.show()
```
